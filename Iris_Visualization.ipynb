# Task 1 - Iris Dataset Visualization

## Objective

The objective of this task is to explore and visualize the Iris dataset using Python libraries such as Pandas, Matplotlib, and Seaborn. This project helps in understanding dataset structure, feature relationships, data distributions, and outliers through different visualization techniques including scatter plots, histograms, and box plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

df['species'] = iris.target
```

```python
df['species'] = df['species'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})
```

```python
print("Shape of dataset:", df.shape)
```

```python
print("Column Names:")
print(df.columns)
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
sns.scatterplot(
    x='sepal length (cm)',
    y='sepal width (cm)',
    hue='species',
    data=df
)

plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df)

plt.title("Box Plot of Iris Features")
plt.show()
```

Conclusion:
